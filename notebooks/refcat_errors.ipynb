```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from refcat_generation import RADecErrors, MagErrors
```

```python
g_mag = np.linspace(10, 22.5, 100)
r_mag = np.linspace(10, 21, 100)
i_mag = np.linspace(10, 21, 100)
```

```python
radec_errors = RADecErrors()
plt.figure(figsize=(10, 5))
ra_err, dec_err = radec_errors(r_mag)
plt.subplot(1, 2, 1)
plt.plot(r_mag, ra_err)
plt.ylim(-0.05e-8, 1e-8)
plt.xlabel("r_mag")
plt.ylabel("RA error")
plt.subplot(1, 2, 2)
plt.plot(r_mag, dec_err)
plt.xlabel("r_mag")
plt.ylabel("Dec error")
plt.ylim(-0.05e-8, 1e-8)
```

```python
plt.figure(figsize=(10, 10))
mag_errors = MagErrors()
for i, (band, mag_values) in enumerate(zip('gri', (g_mag, r_mag, i_mag)), 1):
    plt.subplot(2, 2, i)
    plt.plot(mag_values, mag_errors(mag_values, band))
    plt.ylim(-0.005, 0.05)
    plt.xlabel(f"{band} mag")
    plt.ylabel(f"{band} mag error")
```
